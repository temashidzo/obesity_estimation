The prevalence of obesity has become a significant public health challenge across the globe. With the rise of sedentary lifestyles and high-calorie diets, understanding and addressing the factors contributing to obesity is more critical than ever. The motivation behind this project is to leverage data analysis and predictive modeling to estimate obesity levels among individuals. By doing so, we can gain insights into the patterns and predictors of obesity, which is paramount for developing targeted interventions aimed at mitigating its impact on society.

Obesity is a complex condition influenced by various factors, including genetics, behavior, environment, and culture. It can lead to serious health issues like heart disease, diabetes, and high blood pressure. The problem addressed in this project is the classification of individuals into different obesity levels based on health metrics and lifestyle choices. Accurate classification can aid in early intervention strategies and help healthcare providers offer personalized treatment and advice.

The dataset used in this project is a comprehensive collection of health metrics and lifestyle information from 2111 individuals. It includes data points such as weight, height, age, family history with overweight, food choices, consumption of water, physical activity frequency, and more. The dataset is well-structured, without missing values, covering a wide range of ages and both genders, making it suitable for creating a robust predictive model.

2. Frequancy tables for qualitative response variable and quantitative
explanatory variables. Histograms, summary measures and standard deviations, boxplots, finding the outliers and skewnes for each variable.


```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew
from rich.console import Console

console = Console()

def analyze_variable(df, variable):
    freq=df[variable].value_counts()
    freq=pd.DataFrame({'Variable': freq.keys(), 'Frequency': freq.values})
    freq['Relative Frequency']=(freq['Frequency']/freq['Frequency'].sum()).round(4)*100
    freq=freq.sort_values(by='Relative Frequency', ascending=False)
    freq
    console.print(variable, style='bold green')
    print(f"Frequency table for {variable}:\n{freq.head()}\n")
    
    plt.figure(figsize=(10, 6))
    sns.histplot(df[variable], kde=True)
    plt.title(f'Histogram of {variable}')
    plt.show()
    
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=df[variable])
    plt.title(f'Boxplot of {variable}')
    plt.show()
    
    summary = df[variable].describe()
    print(f"Summary statistics for {variable}:\n{summary}\n")
    print(f"Skewness for {variable}: {skew(df[variable])}\n")
    
variables_to_analyze = ['Weight', 'Age', 'Height', 'NCP', 'CH2O', 'FAF']

for variable in variables_to_analyze:
    analyze_variable(df, variable)
```

4. Multiple linear regression analysis and analys of the coefficients based on

3. Dividing data to train data (70%) and test data (30%).

```python
from sklearn.model_selection import train_test_split
y=df[['NObeyesdad']]
x=df[['Weight', 'Age', 'Height', 'NCP', 'CH2O', 'FAF']]

(x_train , x_test, y_train , y_test) = train_test_split(x,y,test_size=0.3, random_state=0)

train_data=pd.concat([x_train, y_train], axis=1, join='inner')
test_data=pd.concat([x_test, y_test], axis=1, join='inner')
print('Train Data \n',train_data, '\n')

print('Test Data \n',test_data)
```

4. Value counts of test and train data with plots.

```python
import seaborn as sns
import matplotlib.pyplot as plt

print('Train Data - NObeyesdad Count:\n', y_train.value_counts()) 
plt.figure(figsize=(10, 6))
sns.countplot(x='NObeyesdad', data = train_data, palette='hls') 
plt.title('Train Data - NObeyesdad Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print('Test Data - NObeyesdad Count:\n', y_test.value_counts()) 
plt.figure(figsize=(10, 6))
sns.countplot(x='NObeyesdad', data = test_data, palette='hls') 
plt.title('Test Data - NObeyesdad Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

5. Multiple logistic regression analysis based on train data for classification. The predictions (classes) for the test data and comparison of them with the true values.

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(multi_class='multinomial', solver='lbfgs') 

model.fit(x_train, y_train)

yhat = model.predict(x_test)
phat = model.predict_proba(x_test)
p = pd.DataFrame(phat, columns=['p1','p2','p3','p4','p5','p6','p7']).reset_index(drop=True)
pred = pd.DataFrame({'True class':y_test, 'Predicted class': yhat}).reset_index(drop=True) 
result = pd.concat([pred, p], axis=1, join='inner')
print(result)
pred['Predicted class'].value_counts()
```

From the provided results, we observe that out of the total number of patients:

140 are classified as having Obesity Type III.
108 patients are classified as having Insufficient Weight.
95 patients are classified as having Obesity Type II.
86 patients are classified as having Obesity Type I.
85 patients are classified as having Overweight Level I.
66 patients are classified as having Normal Weight.
54 patients are classified as having Overweight Level II.

This distribution suggests that the majority of the patients in this sample are identified with Obesity Type III according to the predictive model used.

6. Confusion matrix and Accuracy value of the fitted model.

```python
from sklearn.metrics import confusion_matrix, accuracy_score

conf_matrix = confusion_matrix(y_test, yhat) 
test_acc = accuracy_score(y_test, yhat)

print('Confusion Matrix: \n', conf_matrix)
print('The accuracy for Test set is {}'.format(round(test_acc*100),3)) 
print('The error rate is {}'.format(round((100-test_acc*100), 3)))
```

The accuracy of the model on the test set is 55%, meaning that the model correctly predicted the class for 55% of the instances in the test set.
The error rate is 45.11%, indicating that 45.11% of the predictions made by the model were incorrect.


These results suggest that while the model has a relatively moderate ability to correctly classify instances, there is still a significant proportion of misclassifications.
